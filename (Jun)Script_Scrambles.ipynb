```python
import random

script_list = []
ans_list = []
title_list = []
title = ['[A]','[B]','[C]']


while True:
    script = str(input())
    splited_script = script.split('.')
    length = len(splited_script)-1

    A, B, C = [], [], []
    a, b, c = 0, 0, 0

    dull_list = []
    for i in range(length):
        dull_list.append(i)


    if (length % 3) == 0:
        criteria  = int(length / 3)
        a = dull_list[:criteria]
        b = dull_list[criteria:2*criteria]
        c = dull_list[2*criteria:]
        
    elif (length % 3) == 1:
        criteria = int((length - 1) / 3)
        a = dull_list[:criteria]
        b = dull_list[criteria:2*criteria]
        c = dull_list[2*criteria:]
        
    else :
        criteria = int((length + 1) / 3)
        a = dull_list[:criteria]
        b = dull_list[criteria:2*criteria]
        c = dull_list[2*criteria:]
    

    for x in a:
        A.append(splited_script[x])
    
    for y in b:
        B.append(splited_script[y])
    
    for z in c:
        C.append(splited_script[z])

    opt_A = ". ".join(A)

    opt_B = ". ".join(B)

    opt_C = ". ".join(C)

    Bulky = [opt_A, opt_B, opt_C]
    index_A = Bulky.index(opt_A)
    index_B = Bulky.index(opt_B)
    index_C = Bulky.index(opt_C)

    The_Last = []
    index = random.sample([0,1,2],3)

    for k in index:
        if k == index_A:
            The_Last.append(Bulky[0])
        elif k == index_B:
            The_Last.append(Bulky[1])
        else:
            The_Last.append(Bulky[2])

    answer = []
        
    
    dic = { 'A' : index[0], 'B' : index[1], 'C' : index[2]}

    for i in range(3):
        if i == dic['A']:
            answer.append('A')
        elif i == dic['B']:
            answer.append('B')
        else:
            answer.append('C')
            
    ans_list.append(answer)
    Fullscript = [The_Last[0], The_Last[1], The_Last[2]]
    Fulltitle = [title[0], title[1], title[2]]
    script_list.append(Fullscript)
    title_list.append(Fulltitle)

    if str(input()) == 'end':
        break
        
print('\n')
for i in range(len(script_list)):
    print('{}번'.format((i+1)))
    print(title_list[i][0], script_list[i][0], '\n')
    print(title_list[i][1], script_list[i][1], '\n')
    print(title_list[i][2], script_list[i][2], '\n')

    
    
for j in range(len(ans_list)):
    print('{}번 : {}'.format((j+1), ans_list[j]))
```
