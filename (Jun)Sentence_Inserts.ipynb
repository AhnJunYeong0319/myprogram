```python
from random import sample 

script_list = []
popped_list = []
ans_list = []
sillies = ['①','②','③','④','⑤']

while True:
    text = str(input())
    splited_text = text.split('.')
    length = len(splited_text) - 1

    if length <= 5:
        print('Error!')
    else :
        lastfiveindicies = [i for i in range(length-5, length)]
        lastfiveindicies
        randindice = int(sample(lastfiveindicies,1)[0])
        randindice

        popped_list.append(splited_text[randindice])
        splited_text.remove(splited_text[randindice])

        for i in range(1,6):
            splited_text[-(i+1)] = splited_text[-(i+1)]+sillies[-i]

        X = '.'.join(splited_text)
        ans = sillies[randindice-(length-5)]
        
        script_list.append(X)
        ans_list.append(ans)
    if str(input()) == 'end':
        break


print('\n')
for i in range(len(script_list)):
    print('-'*77)
    print((i+1),'. ---->   ' + popped_list[i] + '.')
    print('-'*77)
    print('\n'*3)
    print(script_list[i],'\n'*5)
    
    
for j in range(len(ans_list)):
    print('{}번 : {}'.format((j+1), ans_list[j]))
```

```python
import random

script = str(input())
splited_script = script.split('.')
length = len(splited_script)-1

print("\nHow many sentences are in the given script? : {}\n\n\n\n".format(length))
A, B, C = [], [], []
a, b, c = 0, 0, 0

dull_list = []
for i in range(length):
    dull_list.append(i)


if (length % 3) == 0:
    criteria  = int(length / 3)
    a = dull_list[:criteria]
    b = dull_list[criteria:2*criteria]
    c = dull_list[2*criteria:]
        
elif (length % 3) == 1:
    criteria = int((length - 1) / 3)
    a = dull_list[:criteria]
    b = dull_list[criteria:2*criteria]
    c = dull_list[2*criteria:]
        
else :
    criteria = int((length + 1) / 3)
    a = dull_list[:criteria]
    b = dull_list[criteria:2*criteria]
    c = dull_list[2*criteria:]
    

for x in a:
    A.append(splited_script[x])
    
for y in b:
    B.append(splited_script[y])
    
for z in c:
    C.append(splited_script[z])

opt_A = ". ".join(A)

opt_B = ". ".join(B)

opt_C = ". ".join(C)

Bulky = [opt_A, opt_B, opt_C]
index_A = Bulky.index(opt_A)
index_B = Bulky.index(opt_B)
index_C = Bulky.index(opt_C)

The_Last = []
index = random.sample([0,1,2],3)

for k in index:
    if k == index_A:
        The_Last.append(Bulky[0])
    elif k == index_B:
        The_Last.append(Bulky[1])
    else:
        The_Last.append(Bulky[2])


title = ['[A]','[B]','[C]']

print(title[0], The_Last[0], "\n")
print(title[1], The_Last[1], "\n")
print(title[2], The_Last[2], "\n")

answer = []
        
    
dic = { 'A' : index[0], 'B' : index[1], 'C' : index[2]}

for i in range(3):
    if i == dic['A']:
        answer.append('A')
    elif i == dic['B']:
        answer.append('B')
    else:
        answer.append('C')


print(answer)
```
